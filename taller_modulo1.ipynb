### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí

# Promedio
suma = 0
cantidad = 0

for valor in esperanzas_de_vida:
    suma = suma + valor
    cantidad = cantidad + 1

promedio = suma / cantidad


# Máximo
maximo = esperanzas_de_vida[0]

for valor in esperanzas_de_vida:
    if valor > maximo:
        maximo = valor


# Mínimo
minimo = esperanzas_de_vida[0]

for valor in esperanzas_de_vida:
    if valor < minimo:
        minimo = valor

print("Promedio:", promedio)
print("Valor máximo:", maximo)
print("Valor mínimo:", minimo)

````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí
def clasificar_esperanza_vida(anios):
    if anios >= 75:
        return "Alta"
    elif anios >= 60:
        return "Media"
    else:
        return "Baja"

# Pruebas con 5 valores distintos
valores_prueba = [82.4, 75, 68.7, 59.9, 45.2]

for valor in valores_prueba:
    resultado = clasificar_esperanza_vida(valor)
    print("Esperanza de vida:", valor, "-> Clasificación:", resultado)

````

### 1.4 — Listas + Funciones + Bucles (integrador)

**Ejercicio 1.4** — Usa la función creada en 1.3 y la lista del ejercicio 1.1 para:
1. Crear una **nueva lista** llamada `clasificaciones` que contenga la clasificación de cada esperanza de vida
2. Imprimir cada par: `(valor, clasificación)` usando un bucle `for`

````python
# Tu código aquí
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]

def clasificar_esperanza_vida(anios):
    if anios >= 75:
        return "Alta"
    elif anios >= 60:
        return "Media"
    else:
        return "Baja"

# 1. Crear la nueva lista clasificaciones
clasificaciones = []

for valor in esperanzas_de_vida:
    clasificaciones.append(clasificar_esperanza_vida(valor))

# 2. Imprimir cada par (valor, clasificación)
for i in range(len(esperanzas_de_vida)):
    print((esperanzas_de_vida[i], clasificaciones[i]))

````
