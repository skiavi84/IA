### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).  
1. Rellena los valores nulos de columnas **numéricas** con la **media** de cada columna
2. Verifica que ya no queden nulos
3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
# Tu código aquí
import pandas as pd

# Cargar el dataset
df = pd.read_csv("salud_mundial.csv")

# 1. Rellenar los valores nulos de las columnas numéricas con la media
columnas_numericas = df.select_dtypes(include=["float64", "int64"]).columns

for col in columnas_numericas:
    media = df[col].mean()
    df[col] = df[col].fillna(media)

# 2. Verificar que ya no queden valores nulos
print(df.isnull().sum())

````

**Responde aquí** (doble clic para editar):

*¿Por qué usamos la media para rellenar nulos en vez de eliminar las filas?*

> Tu respuesta: Se usó la media para reemplazar los valores nulos en las columnas numéricas y así conservar todas las filas del dataset. Eliminar esas filas implicaría perder información y podría afectar la calidad del análisis. Después del reemplazo, no quedaron valores nulos en ninguna columna.

### 2.3 — Estadísticas descriptivas

🔧 **Ejercicio 2.3**
1. Muestra el resumen estadístico completo del dataset
2. ¿Cuál es la esperanza de vida **promedio** mundial?
3. ¿Cuál es el país con **mayor** gasto en salud? ¿Y el de **menor**?
4. ¿Cuál es la **mediana** de la mortalidad infantil?

````python
# Tu código aquí

import pandas as pd

# Cargar el archivo
df = pd.read_csv("salud_mundial.csv")

# Limpiar nulos como en el ejercicio 2.2
columnas_numericas = df.select_dtypes(include=["float64", "int64"]).columns

for col in columnas_numericas:
    df[col] = df[col].fillna(df[col].mean())

# 1. Resumen estadístico completo
print(df.describe(include="all"))

# 2. Esperanza de vida promedio mundial
promedio_vida = df["esperanza_vida"].mean()
print("Esperanza de vida promedio mundial:", promedio_vida)

# 3. País con mayor y menor gasto en salud
pais_mayor = df.loc[df["gasto_salud_usd"].idxmax(), ["pais", "gasto_salud_usd"]]
pais_menor = df.loc[df["gasto_salud_usd"].idxmin(), ["pais", "gasto_salud_usd"]]

print("Mayor gasto en salud:")
print(pais_mayor)

print("Menor gasto en salud:")
print(pais_menor)

# 4. Mediana de la mortalidad infantil
mediana_mortalidad = df["mortalidad_infantil"].median()
print("Mediana de mortalidad infantil:", mediana_mortalidad)

````

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Tu código aquí
import pandas as pd

# Cargar el dataset
df = pd.read_csv("salud_mundial.csv")

# Limpiar nulos numéricos como en el ejercicio 2.2
columnas_numericas = df.select_dtypes(include=["float64", "int64"]).columns
for col in columnas_numericas:
    df[col] = df[col].fillna(df[col].mean())

# 1. Esperanza de vida promedio por región
vida_promedio_region = df.groupby("region")["esperanza_vida"].mean()
print("Esperanza de vida promedio por región:")
print(vida_promedio_region)

# 2. Gasto en salud promedio por nivel de ingresos
gasto_promedio_ingresos = df.groupby("nivel_ingresos")["gasto_salud_usd"].mean()
print("\nGasto en salud promedio por nivel de ingresos:")
print(gasto_promedio_ingresos)

# 3. Número de países por región
paises_por_region = df.groupby("region")["pais"].count()
print("\nNúmero de países por región:")
print(paises_por_region)

# 4. Región con mayor y menor esperanza de vida promedio
region_mayor = vida_promedio_region.idxmax()
region_menor = vida_promedio_region.idxmin()

print("\nRegión con mayor esperanza de vida promedio:", region_mayor)
print("Región con menor esperanza de vida promedio:", region_menor)
````

### 2.6 — Nueva columna calculada

🔧 **Ejercicio 2.6**
1. Crea una nueva columna llamada `categoria_vida` aplicando la función `clasificar_esperanza_vida` que creaste en la Parte 1 a la columna `esperanza_vida`
2. ¿Cuántos países caen en cada categoría? (usa `value_counts()`)
3. ¿Qué porcentaje del total representa cada categoría?

````python
# Tu código aquí
import pandas as pd

# Cargar el dataset
df = pd.read_csv("salud_mundial.csv")

# Rellenar nulos numéricos con la media
columnas_numericas = df.select_dtypes(include=["float64", "int64"]).columns
for col in columnas_numericas:
    df[col] = df[col].fillna(df[col].mean())

# Función creada en la Parte 1
def clasificar_esperanza_vida(anios):
    if anios >= 75:
        return "Alta"
    elif anios >= 60:
        return "Media"
    else:
        return "Baja"

# 1. Crear nueva columna
df["categoria_vida"] = df["esperanza_vida"].apply(clasificar_esperanza_vida)

# 2. Cantidad de países por categoría
conteo_categorias = df["categoria_vida"].value_counts()
print("Cantidad por categoría:")
print(conteo_categorias)

# 3. Porcentaje por categoría
porcentaje_categorias = df["categoria_vida"].value_counts(normalize=True) * 100
print("\nPorcentaje por categoría:")
print(porcentaje_categorias)

````

**¿Qué archivos debería incluir un `.gitignore` en un proyecto de Data Science? Menciona al menos 3 y explica por qué.**

> Tu respuesta: En un proyecto de Data Science, el archivo .gitignore debe incluir elementos que no es necesario versionar.
